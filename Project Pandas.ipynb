```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

read_file = pd.read_excel ('Seoul_apt_data.xlsx')
read_file.to_csv ('Seoul_apt_data.csv', index = None, header=True)
data = pd.read_csv('Seoul_apt_data.csv')
```

```python
write_sigu_max = data.groupby('구').max()
write_sigu_max.to_csv('sigu_max.csv')
write_sigu_max
```

```python
write_sigu_min = data.groupby('구').min()
write_sigu_min.to_csv('sigu_min.csv')
write_sigu_min
```

```python
write_dong_min = data.groupby('동').min()
write_dong_min.to_csv('dong_min.csv')
write_dong_min
```

```python
write_dong_max = data.groupby('동').max()
write_dong_max.to_csv('dong_max.csv')
write_dong_max
```

```python
data= pd.read_csv("sigu_max.csv",encoding='utf-8')

sns.barplot(data = data, x="가격", y="구")

plt.show()
```

```python
data= pd.read_csv("sigu_min.csv",encoding='utf-8')

sns.barplot(data = data, x="가격", y="구")

plt.show()
```

```python
write_area_mean = data.groupby('구').mean(['가격'])
write_area_mean.to_csv('area_mean.csv')
write_area_mean
```

```python
write_mean = data.groupby('구').mean(['가격'])
write_mean.to_csv('sigu_mean.csv')
write_mean
```

```python
data= pd.read_csv("sigu_mean.csv",encoding='utf-8')

sns.barplot(data = data, x="가격", y="구")

plt.show()
```
